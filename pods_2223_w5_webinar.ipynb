```python
import pandas as pd
```

```python
dataset_location = 'https://raw.githubusercontent.com/gerberl/6G7V0026-2223/main/datasets/gapminder.tsv'
gm = pd.read_csv(
    dataset_location,
    sep='\t'
)
```

```python
gm.head()
```

```python
(gm
 .loc[ gm['year']==2007 ]
 .set_index('country')
).head()
```

```python
gm_2007 = (gm
 .loc[ gm['year']==2007 ]
 .set_index('country')
)
```

```python
gm_2007.head(1)
```

```python
gm_2007.loc[ ['Brazil', 'India', 'Nigeria'] ]
```

```python
gm_2007_top5 = gm_2007.sort_values('pop', ascending=False).head(5)
```

```python
gm_2007_top5
```

```python
total_pop = gm_2007['pop'].sum()
```

```python
# vectorisation: apply an operation to a vector at an element-wise level
# index alignment: if working with Series or DataFrame, rows/columns/values are
# aligned for the computation based on their index values
gm_2007_top5['pop'] / total_pop * 100
```

```python
gm_2007_top5['pop_perc'] = gm_2007_top5['pop'] / total_pop * 100
```

```python
# bear in mind the concepts of views and copies
# if you take a slice of DataFrame and try to assign a column to it
# you might modifying the original DataFrame
# you can always take a copy of DataFrame/Series with .copy() if you 
# want to perform some experiments on it without messing it up
# https://jorisvandenbossche.github.io/blog/2022/04/07/pandas-copy-views/
```

```python
gm_2007_top5
```

```python
(gm_2007_top5['pop'] * gm_2007_top5['gdpPercap']) / 1000000000
```

```python
gdp = (gm_2007_top5['pop'] * gm_2007_top5['gdpPercap'] / 1000000000).loc['China':'Indonesia']
```

```python
# gm_2007_top5['gdp'] = gdp
```

```python
gm_2007_top5
```

```python
# joining DataFrames: contact, merge (join)
# https://jakevdp.github.io/PythonDataScienceHandbook/03.07-merge-and-join.html
pd.concat( [ gm_2007_top5, gdp.rename('gdp') ], axis='columns' )
```

```python
china = gm.loc[ gm['country']=='China' ].set_index('year')
```

```python
china.head(3)
```

```python
india = gm.loc[ gm['country']=='India' ].set_index('year')
```

```python
india.head(3)
```

```python
china['lifeExp'] - india['lifeExp']
```

```python
china_india_diff = china['lifeExp'] - india['lifeExp']
```

```python
india_china_life_exp = pd.concat(
    [ china['lifeExp'], india['lifeExp'], china_india_diff ], axis='columns'
)
```

```python
# india_china_life_exp.set_axis([ 'China_LE', 'India_LE', 'LE_Diff' ], axis='columns')
pd.concat(
    [ 
        china['lifeExp'].rename('China_LE'), 
        india['lifeExp'].rename('India_LE'), 
        china_india_diff.rename('LE_Diff')
    ], axis='columns'
)
```

```python
# using vectorised operations is normally much more efficient than
# writing loops over rows or columns
# reasons? there are optimisations in place in the `pandas`/`numpy` engine
# also, more concise/elegant/simple
```
